```python
# pandas liberaries is uesd to import the value from database, table etc
import pandas as ps
```

```python
# store the Salary_Data.csv file into dataset variable with use of read_csv function 
dataset=ps.read_csv("Salary_Data.csv")
dataset
```

```python
# assign and store input value (X) into independent variabe
independent=dataset[["YearsExperience"]]
independent
```

```python
#assign and store ouput value(Y) into dependent variabe
dependent=dataset[["Salary"]]
dependent
```

```python
# Here we spliting train and test set with the use of sklearn.model_selection & train_test_split function
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
#creating model use of fit linear regression function and passing the training set input and output value
regressor.fit(x_train,y_train)
```

```python
# see the value of Weight and Bais
weight=regressor.coef_
bais=regressor.intercept_
weight
bais
```

```python
# to Predcit model , we are passing test set input value 
y_pred=regressor.predict(x_test)
```

```python
# using r2 score function to validate the our model is good or poor, if r score value comes near to 1 then its a good model and r2 score value comes under 0 then its a poor model
from sklearn.metrics import r2_score
```

```python
# passing actual output and predict output
r_score=r2_score(y_test,y_pred)
r_score
```

```python
# using pickle to store the best model and extension should be .sav , wb- write binaries , rb- read binaries
import pickle
filename="finalized_model_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
# load function is used to load the saved model.
load_model=pickle.load(open("finalized_model_linear.sav",'rb'))
ye=int(input("Enter YearsExperience:"))
result=load_model.predict([[ye]])
result
```
